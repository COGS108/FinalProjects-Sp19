```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
top_songs=pd.read_csv('song.csv')
```

```python
# data to plot
n_groups = 10

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8
rects1 = plt.bar(index, top_songs['danceability'].head(10), bar_width,
alpha=opacity,
label='danceability')

rects2 = plt.bar(index + bar_width, top_songs['energy'].head(10), bar_width,
alpha=opacity,
label='energy')

plt.xlabel('Top songs')
plt.ylabel('Trait level')
plt.title('Top songs and their musical attributes')
plt.xticks(index + bar_width, top_songs['name'].head(10))
plt.legend()
plt.xticks(rotation=90)
plt.show()
```
