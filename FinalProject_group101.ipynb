# Research Question

QUESTION: HOW DOES AN INDIVIDUAL'S INCOME AFFECT THEIR NUTRITIONAL HEALTH?

For our research we wanted to see if there is a relationship between income and health. For income we looked at the ratio of income to the poverty level (a dollar amount set by the government). This value ranges from 0 to 5, which means an income of 0 times the poverty level to an income of 5 times the poverty level. Values above 5 are topcoded as 5. We decided to divide the range into two categories of 0-1 (below the poverty line) and 1-5 (above the poverty line). To measure health, we looked examination dataset of Body Mass Index(BMI) and blood pressure, and we looked at lab measurements of Cholesterol, Triglycerides, Insulin. We also looked at a questionnaire with Yes or No questions about the individuals’ history of health problems such as diabetes, heart problems, and thyroid problems. We then compared every health variable with the two-category variable of income to see if the health level between people below and above the poverty line is significantly different or not.

```python
# Imports
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression

import seaborn as sns; sns.set(color_codes=True)
import patsy
import statsmodels.api as sm
import scipy.stats as stats
from scipy.stats import ttest_ind, chisquare, normaltest

#load Datasets
demo = pd.read_csv('demographic.csv')
exam = pd.read_csv('examination.csv')
quest = pd.read_csv('questionnaire.csv')
labs = pd.read_csv('labs.csv')
```

# Data Cleaning

First, we isolated the columns we chose to analyze (as listed in Dataset) and dropped all nan values in those columns. Also, in accordance with Safe Harbour, we drop the data for the ages below 18.

```python
#Demographic: drop unneccessary columns and NaNs
demo = demo.filter(['SEQN','INDFMPIR','RIDAGEYR'])
demo = demo[demo['RIDAGEYR']>=18]
demo = demo.dropna()

#Examination: drop unneccessary columns and NaNs
exam = exam.filter(['SEQN','BPXDI1','BPXDI2','BPXDI3','BPXSY1', 
                    'BPXSY2','BPXSY3','BMXBMI','BMXWT','BMXHT'])
exam = exam.dropna()

#Questionaire: drop unneccessary columns and NaNs
quest = quest.filter(['SEQN','BPQ020','BPQ080','DIQ010','MCQ080',
                      'MCQ160B','MCQ160C','MCQ160E','MCQ160M'])
quest = quest.dropna()

#Labs: drop unneccessary columns and NaNs
labs = labs.filter(["SEQN","LBDHDDSI","LBDLDLSI","LBDINSI","LBDTRSI"])
labs = labs.dropna()
```

All of the column names were changed from the original code/number combinations to more reflective names of the type of data in the columns. This gave us a better understanding of what was in the dataframe, without having to refer to the code dictionary that was provided with the original datasets. Finally, we also created three different datasets to analyze, merging each of the other datasets to the demographics dataset.

```python
#demographic and examination merge, column rename for clarity
df = demo.merge(exam, on='SEQN')
df.rename(index=str, columns={"RIDAGEYR":"age",
                              "INDFMPIR":"income_ratio",
                              "BPXDI1":"diastolic_1",
                              "BPXDI2":"diastolic_2",
                              "BPXDI3":"diastolic_3",
                              "BPXSY1":"systolic_1",
                              "BPXSY2":"systolic_2",
                              "BPXSY3":"systolic_3",                                   
                              "BMXBMI":"bmi",
                             },
         inplace = True)

#demographic and questionaire merge, column rename for clarity
joint = demo.merge(quest, on='SEQN')
joint.rename(columns = {'INDFMPIR': 'income_ratio',
                        'BPQ020': 'high_blood_pressure',
                        'BPQ080': 'high_cholesterol',
                        'DIQ010': 'diabetes',
                        'MCQ080': 'overweight',
                        'MCQ160B': 'congestive_heart_failure',
                        'MCQ160C': 'coronary_heart_disease',
                        'MCQ160E': 'heart_attack',
                        'MCQ160M': 'thyroid_problem'},
            inplace = True)

#demographic and labs merge, column rename for clarity
combined = demo.merge(labs, on="SEQN")
combined.rename(index=str, columns={ 
                            "INDFMPIR":"income_ratio", 
                            "RIDAGEYR" :"age",
                            "LBDHDDSI":"HDL",
                            "LBDLDLSI":"LDL",                                            
                            "LBDINSI":"insulin",                                      
                            "LBDTRSI":"triglyceride"},
               inplace=True)
```

Here is the data visualization, analysis, and results obtained from the questionnaire dataset combined with the demographic dataset. For the columns from the questionnaire dataset, each column represents a question asking whether the participant was ever medically diagnosed with the condition or not. The response to the question is represented by 1 meaning yes, 2 meaning no, 3 meaning the participant didn't know or weren't sure, and 9 meaning the participant chose not to respond. We classified participants has high or low income by their income ratio. If their ratio of income to the poverty guidline was less than one they were classified as having a low income and if their ratio was greater than or equal to one, they were classified as having a high income. We now wanted to visualize how many people did or didn't have each condition separately depending on whether they had a high or low income. Hence, we wanted to further clean this dataframe in order to remove any rows that had the values of 3 or 9. We created 2 different dataframes, one having values of the column of the condition when the participants had a low income and the other when the participants had a high income. The following code was for the column asking about high blood pressure. After creating the new dataframes and dropping the rows with values of 3 or 9, we created 2 more dataframes with the count of the remaining values, 1 and 2.

```python
bpp = joint['high_blood_pressure'].loc[joint['income_ratio'] < 1]
for index, value in bpp.iteritems():
    if value == 3 or value == 9:
        bpp = bpp.drop(labels = index)
bpp_vc = bpp.value_counts()
        
bpr = joint['high_blood_pressure'].loc[joint['income_ratio'] >= 1]
for index, value in bpr.iteritems():
    if value == 3 or value == 9:
        bpr = bpr.drop(labels = index)
bpr_vc = bpr.value_counts()
```

We then combined both the high-income and low-income value-count dataframes, renamed the 1 values as Yes and 2 values as No, and bar-plotted the resulting dataframe as subplots for each income level.

```python
bp_df = pd.DataFrame({'Low Income': bpp_vc,'High Income': bpr_vc})
bp_df.rename(index={2.0: 'No', 1.0: 'Yes'}, inplace = True)
bp_df.plot.bar(rot=0, subplots = True)
plt.xlabel('Response to question')
plt.ylabel('No. of people diagnosed with high blood pressure')
plt.show()
```

Visualizing this in such a way led us to wanting to compare the ratio of No responses to Yes responses between each income level and to see if there is a significant difference between the two. Hence, we performed the chi-squared test because the variables and conditions we wanted to analyze were categorical. If the p-value from the chi-squared test was less than the significance level of 0.05, the null hypothesis was rejected and there was a significant difference between the ratio of responses in each income level. For high blood pressure, we found that there was no significant difference between having a high or low income and being diagnosed with high blood pressure

```python
chi1, p_bp, dof1, expected1 = stats.chi2_contingency(np.array([bpp_vc.values, bpr_vc.values]))
if p_bp < 0.05:
    print("There is a significant difference!")
else:
    print("There is no significant difference")
```

We performed similar cleaning, visualization, and chi-squared test analysis on the rest of the conditions. The following is for being diagnosed with high cholesterol. We removed values of 3 or 9, divided the dataset by high or low income, and created dataframes with the count of the remaining values for each income level.

```python
cp = joint['high_cholesterol'].loc[joint['income_ratio'] < 1]
for index, value in cp.iteritems():
    if value == 3 or value == 9:
        cp = cp.drop(labels = index)
cp_vc = cp.value_counts()
        
cr = joint['high_cholesterol'].loc[joint['income_ratio'] >= 1]
for index, value in cr.iteritems():
    if value == 3 or value == 9:
        cr = cr.drop(labels = index)
cr_vc = cr.value_counts()
```

We then again combined the new value-count dataframes, renamed the 1 and 2 values, and bar-plotted the resulting dataframe by each income level.

```python
chol_df = pd.DataFrame({'Low Income': cp_vc,'High Income': cr_vc})
chol_df.rename(index={2.0: 'No', 1.0: 'Yes'}, inplace = True)
chol_df.plot.bar(rot=0, subplots = True)
plt.xlabel('Response to question')
plt.ylabel('No. of people diagnosed with high cholesterol')
plt.show()
```

On performing the chi-squared test, we found that there was a significant difference between having a high or low income and being diagnosed with high cholesterol! By looking at the bar plot, we can see that the ratio between people who have and don't have high cholesterol is smaller among people with higher income level compared to people with lower income level.

```python
chi2, p_chol, dof2, expected2 = stats.chi2_contingency(np.array([cp_vc.values, cr_vc.values]))

if p_chol < 0.05:
    print("There is a significant difference!")
else:
    print("There is no significant difference")
```

With that, we will first look at the HDL data, printing a histogram of it. From this, we can see that the distribution of HDL is normal, perhaps with very slight positive skew. Then, we create a scatter of the HDL data against the income ratio to the poverty line, along with the regression line of the scatter data. Thus, while the slope is small, there is a statistically significant correlation between income ratio and HDL levels.

```python
combined.hist('HDL', bins=25)
sns.lmplot(x="income_ratio", y="HDL", data=combined, line_kws={'color': 'orange'})

slope, intercept, r_value, p_value, std_err = stats.linregress(combined["income_ratio"],
                                                               combined["HDL"])
print("p value:",p_value)
print("Does income ratio predict HDL?", p_value <= 0.05)
```

We can further observe the difference of HDL of those above and below the poverty line of 1. By performing a chi-square test between the those two groups, we can see that the difference between them, while appearing minor, is still significant.

```python
#p test for HDL poverty vs high income
poverty = combined["HDL"].loc[combined["income_ratio"] < 1]
highinc = combined["HDL"].loc[combined["income_ratio"] >= 1]

t_val, p_val = stats.ttest_ind(poverty,highinc)
print("p value:",p_val)

if p_val < 0.05:
    print('There is a significant difference!')
else:
    print('There is NOT a significant difference!')
```

We explore LDL in the same fashion, observing that the histogram is also largely normal with a slight positive skew. The p-value generated for this scatter plot also falls under our critical region of .05, thus also indicating a correlation of LDL with income ratio. 

```python
combined.hist('LDL', bins=25)
sns.lmplot(x="income_ratio", y="LDL", data=combined, line_kws={'color': 'orange'})

slope, intercept, r_value, p_value, std_err = stats.linregress(combined["income_ratio"],
                                                               combined["LDL"])
print("p value:",p_value,)
print("Does income ratio predict LDL?", p_value <= 0.05)
```

In contrast, when observing the difference of LDL of those above versus those below the poverty line with a chi-square test, there is no significant difference shown. This provides an interesting conundrum, as the previous p-test indicates that there should be an overall trend of change as income ratio increases. The result for this test implies that either the previous test was faulty, falling within the 1% change indicating there was a correlation when there was none, or that the poverty line is not the critical breakpoint of where LDL levels shift.

```python
#p test for LDL poverty vs high income
poverty = combined["LDL"].loc[combined["income_ratio"] < 1]
highinc = combined["LDL"].loc[combined["income_ratio"] >= 1]

t_val, p_val = stats.ttest_ind(poverty,highinc)
print("p value:",p_val)

if p_val < 0.05:
    print('There is a significant difference!')
else:
    print('There is NOT a significant difference!')
```

As we see, if we instead compare those above and below the midpoint of the data, we can see a significant difference there.

```python
#p test for LDL poverty vs high income
poverty = combined["LDL"].loc[combined["income_ratio"] < 2.5]
highinc = combined["LDL"].loc[combined["income_ratio"] >= 2.5]

t_val, p_val = stats.ttest_ind(poverty,highinc)
print("p value:",p_val)

if p_val < 0.05:
    print('There is a significant difference!')
else:
    print('There is NOT a significant difference!')
```

Due to prior research, we are aware that insulin levels drop with age. Thus, there is a possibility that the correlation between income ratio and insulin is due to age instead. Thus, we perform the following patsy.dmatrices with the formula insulin ~ income_ratio + age to check whether the relationship between insulin and income ratio is still observable accounting for age. As the p-value falls belows our critical value, it is determined that relationship still exists.

```python
#linear model for insulin to income ratio and age
out, pred = patsy.dmatrices('insulin ~ income_ratio + age', combined)
mod = sm.OLS(out, pred)
type(mod)
res = mod.fit()

#Does income ratio still predict insulin?
print("p value:",res.pvalues[1])
print("Given age, does income ratio still predict insulin?")
print(res.pvalues[1] <= 0.05)
```

However, like the LDL test, there is no statistically significant difference in insulin levels of those below the poverty level and those above the poverty level. And like the conclusion for the LDL test, either the previous linear regression test was faulty, or the poverty line is not the critical breakpoint of where insulin levels shift.

```python
#p test for insulin poverty vs high income
poverty = combined["insulin"].loc[combined["income_ratio"] < 1]
highinc = combined["insulin"].loc[combined["income_ratio"] >= 1]

t_val, p_val = stats.ttest_ind(poverty,highinc)
print("p value:",p_val)

if p_val < 0.05:
    print('There is a significant difference!')
else:
    print('There is NOT a significant difference!')
```

This can be observed if we instead compare those above and below 2.5 times the poverty level, of which there is a significant difference of insulin between.

```python
#p test for insulin poverty vs high income
poverty = combined["insulin"].loc[combined["income_ratio"] < 2.5]
highinc = combined["insulin"].loc[combined["income_ratio"] >= 2.5]

t_val, p_val = stats.ttest_ind(poverty,highinc)
print("p value:",p_val)

if p_val < 0.05:
    print('There is a significant difference!')
else:
    print('There is NOT a significant difference!')
```
