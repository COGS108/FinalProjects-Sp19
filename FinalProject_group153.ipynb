#### Discover and Remove Outliers
We want to remove outliers from our dataset that could have been
put there by clinical error. We will use the 1.5 x IQR rule to detect and remove outliers.

```python
def plotDists(df, labels, structnum):
    plt.subplots_adjust(bottom=-1, right=1.5, wspace=0.5, hspace=0.5)
    
    
    for i, label in zip(range(1, len(labels)+1), labels):
        plt.subplot(structnum+i)
        sns.distplot(df[label])
```

```python
# detect outliers that are more than 1.5 IQR off the 
# 25th and 75th percentile
def show_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 # Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    
    df_removed = df_in[(df_in[col_name] <= fence_low) | 
                       (df_in[col_name] >= fence_high)]    
    
    return  df_removed
```

```python
def remove_outliers(df_in, col_name, top_q=False):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 # Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr

    if top_q:
        df_out = df_in.loc[(df_in[col_name] < fence_high)]
    else:
        df_out = df_in.loc[(df_in[col_name] > fence_low)]

    return  df_out
```

##### Go through columns, and discover what outliers exist

```python
# for our 4 numerical values in this dataset, we want to remove outliers
# let's see what outliers we detect

def outlierDetection(df, numerical_cat):
    for col in numerical_cat:
        outliers = show_outlier(df, col)
        print('found ', len(outliers), ' in column ', col)

        if (len(outliers) > 0):
            plt.title(col + ' distribution in data')
            s = outliers[col].sort_values(ascending=False).values
            
            density = gaussian_kde(df[col])
            xs = np.linspace(min(df[col]), max(df[col]), 200)
            
            density.covariance_factor = lambda: .25
            density._compute_covariance()
            
            plt.plot(xs, density(xs))
            plt.show()

            print ('total number of outliers', len(outliers))
            print ('top ten outliers', s[:10])
            print ('bot ten outliers', s[-10:])
                    
        print ('==============================')
```

```python
# Our numerical features are: age, systolic_bp, chol, max_heart_rate
# Let's see their box plots

def makeBoxPlots(df, cols):
        
    fig, axs = plt.subplots(len(cols), 1)
    for i, col in enumerate(cols):
                
        axs[i].boxplot(df[col], 0, 'rs', 0, showmeans=True, widths=0.75)
        axs[i].set_title(col)

    
    fig.subplots_adjust(left=2, right=5, bottom=-1, top=2,
                    hspace=0.4, wspace=0.3)
```

__Pulse Pressure can be an indicator of heart disease:__ https://www.medhelp.org/posts/Heart-Disease/Pulse-Pressure-As-An-Indicator-of-Heart-Disease/show/1576075


We were able to use patients’ systolic and diastolic blood pressures in order to calculate their pulse pressure (the difference between systolic and diastolic). The pulse pressure (whether low or high) can also tell us if a patient is more likely to heart disease. According to sources at the Mayo Clinic, “For adults older than age 60, a pulse pressure greater than 60 can be a useful predictor of heart attacks or other cardiovascular disease; this is especially true for men”. We decided to include pulse pressure in our data to detect abnormally low or high pulse pressures in patients then assign them a 0 or 1 depending on if their pulse pressure indicated potential chances of affected heart health (if it’s too high or too low).

```python
hos_data.insert(0,'pulse_pressure', 
                hos_data['systolic_bp'] - hos_data['diastolic_bp'])
hos_data['pulse_pressure'].hist()
```

__BMI can be an indicator of heart health__: https://www.nhlbi.nih.gov/health/educational/lose_wt/risk.htm

```python
# bmi = weight in kilograms / (height in meters squared)
hos_data.insert(0, 'bmi', hos_data['weight'] / 
                (hos_data['height'] * hos_data['height']/10000 ))
hos_data['bmi'].hist()
```

```python
def plotAbsCorrMatrix(dat):
    sns.set(style='white')
    corr = np.absolute(dat.corr())

    # Generate a mask for the upper triangle
    mask = np.zeros_like(corr, dtype=np.bool)
    mask[np.triu_indices_from(mask)] = True

    # Set up the matplotlib figure
    f, ax = plt.subplots(figsize=(11, 9))

    # Generate a custom diverging colormap
    cmap = sns.diverging_palette(220, 10, as_cmap=True)

    # Draw the heatmap with the mask and correct aspect ratio
    sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
                square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

# Model Building

```python
# source: https://scikit-learn.org/stable/auto_examples/
# classification/plot_classifier_comparison.html
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_moons, make_circles, make_classification
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

```python
# code from https://scikit-learn.org/stable/auto_examples/
# model_selection/plot_confusion_matrix.html
from sklearn.metrics import confusion_matrix
from sklearn.utils.multiclass import unique_labels

def plot_confusion_matrix(y_true, y_pred,
                          normalize=False,
                          title=None,
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if not title:
        if normalize:
            title = 'Normalized confusion matrix'
        else:
            title = 'Confusion matrix, without normalization'

    # Compute confusion matrix
    cm = confusion_matrix(y_true, y_pred)
   
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]     

    fig, ax = plt.subplots()
    im = ax.imshow(cm, interpolation='nearest', cmap=cmap)
    ax.figure.colorbar(im, ax=ax)
    # We want to show all ticks...
    ax.set(xticks=np.arange(cm.shape[1]),
           yticks=np.arange(cm.shape[0]),
           # ... and label them with the respective list entries
           title=title,
           ylabel='True label',
           xlabel='Predicted label')

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
             rotation_mode="anchor")

    # Loop over data dimensions and create text annotations.
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, format(cm[i, j], fmt),
                    ha="center", va="center",
                    color="white" if cm[i, j] > thresh else "black")
    fig.tight_layout()
    return ax
```

```python
names = ["Nearest Neighbors", "Linear SVM", "RBF SVM",
         "Random Forest", "Neural Net", "AdaBoost",
        "Naive Bayes"]

classifiers = [
    KNeighborsClassifier(2),
    SVC(kernel="linear", C=0.025),
    SVC(gamma=2, C=1),
    RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),
    MLPClassifier(alpha=1, max_iter=1000),
    AdaBoostClassifier(),
    GaussianNB(),
]

zipNmClf = zip(names, classifiers)
# zipNmData = zip(
#     ('uci_male', 'uci_female', 'hos_male', 'hos_female'),
#     (uci_male, uci_female, hos_male, hos_female)
# )
```

```python
def clfResults(dataset, dataset_name, zipNmClf):
    
    classifierEvals = dict()

    train, test = train_test_split(dataset, test_size=0.2)
    
    X_train = train.drop('target', axis=1).values.tolist()
    Y_train = train['target'].apply(int).values.tolist()
    
    X_test = test.drop('target', axis=1).values.tolist()
    Y_test = test['target'].apply(int).values.tolist()
    
    for clf_name, clf in zip(names, classifiers):
        clf.fit(X_train,Y_train)
        
        predicted = clf.predict(X_test)
        acc = accuracy_score(Y_test, predicted)
        
        title = clf_name + ': ' + dataset_name + ' - ' 
        + str(acc)[:4]
        
        ax = plot_confusion_matrix(Y_test, predicted,
                             normalize=True,
                             title=title)
    
        classifierEvals[title] = (acc, ax)
        
    return classifierEvals
```
