I was given a data set that in its raw form was a file with a bunch of pdf files. If there were an updated file, I deleted the non-updated one since i was given both.

```python
import os
from PyPDF2 import PdfFileMerger
```

```python
years = [y for y in os.listdir("Data") if y != '.DS_Store']
# Creates the Months; first iteration
for y in years:
    months = [m for m in os.listdir(f"Data/{y}") if m != '.DS_Store']
    for m in months:
        if os.path.isfile(f'./FirstIteration/{y}/{m}.pdf'):
            print(f"File Exists: {m}")
            continue
        merger = PdfFileMerger()
        days = [d for d in os.listdir(f"Data/{y}/{m}") if d != '.DS_Store']
        # July 4 is the day of split, will move July 1, 2, 3, 4 to Jun Folder
        for d in days:
            merger.append(open(f"Data/{y}/{m}/{d}", "rb"))
        #with open(f"{y},{m}.pdf", "w") as fout:
        merger.write(f"FirstIteration/{y}/{m}.pdf")
        merger.close()
        print(f"Success: {m}")
```

```python
# Second Iteration
fi = [y for y in os.listdir("FirstIteration") if y != '.DS_Store']
syntax2 = ["APRIL 2011.pdf", "FEBRUARY 2011.pdf", "JANUARY 2011.pdf", 
           "JUNE 2011.pdf", "MARCH 2011.pdf", "MAY 2011.pdf"]
merger2 = PdfFileMerger()


for y in fi:
    if os.path.isfile(f'./SecondIteration/{y}.pdf'):
            print(f"File Exists: {y}")
            continue
    merger = PdfFileMerger()
    merger2 = PdfFileMerger()
    months = [m for m in os.listdir(f"FirstIteration/{y}") if m != '.DS_Store']
    for m in months:
        # April 2011 has differing syntax; be sure to take that into account
        if m in syntax2:
            if os.path.isfile(f'./SecondIteration/{y}_2.pdf'):
                print(f"File Exists: {y}_2")
                continue
            merger2.append(open(f"FirstIteration/{y}/{m}", "rb"))
        else:
            merger.append(open(f"FirstIteration/{y}/{m}", "rb"))
    merger.write(f"SecondIteration/{y}.pdf")
    merger.close()
    print(f"Success: {y}")
merger2.write(f"SecondIteration/2011_2.pdf")
merger2.close()
```
