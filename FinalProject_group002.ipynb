```python
import pandas as pd 

dropout = pd.read_csv('Kidsdata-Students-Not-Completing-High-School-2010-2015.csv') 
dropout = dropout.dropna(axis = 1, how = 'all')

dropout = dropout.dropna(axis = 0, how = 'all')
dropout = dropout.iloc[:-4]
dropout = dropout.fillna(0)
dropout = dropout.rename(index=str, 
                         columns={"Unnamed: 0": "Location", 
                                  "Unnamed: 1": "2010", 
                                  "Unnamed: 2": "2011",
                                  "Unnamed: 3": "2012",
                                  "Unnamed: 4": "2013", 
                                  "Unnamed: 5": "2014",
                                  "Unnamed: 6": "2015" })
dropout = dropout.iloc[3:]


##replacing unnecesarry characters in the datasets

dropout["2010"] = dropout["2010"].str.replace(",","")
dropout["2011"] = dropout["2011"].str.replace(",","")
dropout["2012"] = dropout["2012"].str.replace(",","")
dropout["2013"] = dropout["2013"].str.replace(",","")
dropout["2014"] = dropout["2014"].str.replace(",","")
dropout["2015"] = dropout["2015"].str.replace(",","")
dropout["2010"] = dropout["2010"].str.replace("%","")
dropout["2011"] = dropout["2011"].str.replace("%","")
dropout["2012"] = dropout["2012"].str.replace("%","")
dropout["2013"] = dropout["2013"].str.replace("%","")
dropout["2014"] = dropout["2014"].str.replace("%","")
dropout["2015"] = dropout["2015"].str.replace("%","")

dropout['2010'] = pd.to_numeric(dropout['2010'],errors='coerce')
dropout['2011'] = pd.to_numeric(dropout['2011'],errors='coerce')
dropout['2012'] = pd.to_numeric(dropout['2012'],errors='coerce')
dropout['2013'] = pd.to_numeric(dropout['2013'],errors='coerce')
dropout['2014'] = pd.to_numeric(dropout['2014'],errors='coerce')
dropout['2015'] = pd.to_numeric(dropout['2015'],errors='coerce')



#take average of each row 
dropout['average'] = dropout.sum(axis=1)

dropout = dropout.iloc[1:]
```

## Privacy/Ethics Considerations
The datasets used were provided publicly on the data platform, Kaggle, or came from open source websites, ensuring that these were public platforms with public data accessible by any individual. According the Safe Harbour rules, no privacy concerns arise about these datasets. Because of this, we believe there is no concern about privacy. There was no personal information included in any of these datasets, instead the data are all simple count totals of various rates of homelessness or drop-out rates, so there is no concern of personal information. All data was anonymous.


We acknowledge that any correlation in our data analysis of homelessness and drop-out rates doesn't necessarily equate causation. Those that drop out may not necessarily become homeless. There are many other factors that can cause a student to drop out that are unaccounted for in our research projects such as health concerns, poverty, special education needs, and other issues. There are also many other factors that can cause an individual to be homeless such as poverty, drugs, life circumstances, unsupportive government programs, and capitalism. 


We also recognize that our data is not comprehensive or holistic in measuring the actual rates of homelessness or drop-out rates. Many people are historically not measured in these surveys and dataset, and thus, the actual number of people experiencing these circumstances may be much higher. Similarly, we acknowledge that there may be potential biases that could have occurred in data collection, thereby also skewing the rates delineated. There may be many different methodologies in determining and classifying a homeless individual. The datasets we found did not provide the methodologies they used to collect the data. Because of this, we acknowledge that bias in counting who is homeless and a drop-out could have occurred and thus, the data and results could potentially be skewed based on the biases of those collecting the data.
