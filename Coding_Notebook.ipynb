### HELPER FUNCTIONS
Our data is very unorganized, so we make use of several helper functions detailed below

```python
"""
NOTE: this only works for fatality and structural columns

column_separator(df, col, new_names)

Separates one column into two (if values are separated by spaces)

params:
df : DataFrame to change
col : column name to split
new_names : array of 2 new names for columns

returns:
pandas DataFrame with changes applied
"""
def column_separator(df, col, new_names):
    column = df[col].astype(str).apply(lambda x : x.split(' '))
    left_arr = []
    right_arr = []
    #uncomment below to see actual array values
    #display(column)
    
    #iterate through values to separate them
    for val in column:
        #edge case array is len 1
        if len(val) == 1:
            left_arr.append(np.nan)
            right_arr.append(np.nan)
        #normal case array len 2
        elif len(val) == 2:
            left_arr.append(val[0])
            right_arr.append(val[1])
        #edge case array is bigger than expected
        else:
            left_arr.append('error')
            right_arr.append('error')
    #now make new column names and remove the old one
    #display(left_arr)
    #display(right_arr)
    ans_df = df.drop(col, axis=1)
    ans_df[new_names[0]] = left_arr
    ans_df[new_names[1]] = right_arr
    return ans_df
"""
Separates one column into two columns, values must be separated by spaces; drops the original column at iloc index

params:

df : the dataframe to modify
index : the index of the column as given by iloc indexing (int)
new_names : the new names for the two new columns created (array size 2)
front : determines whether to separate the value from the front or from the back (boolean, optional):

Example:
Consider the following string:
Apple Cinnamon Bannana

front = True will give
[Apple] [Cinnamon Bannana]

front = False will give
[Apple Cinnamon] [Bannana]

returns:
pandas DataFrame with changes applied
"""

def robust_col_sep(df, index, new_names, front=True):
    column = df.iloc[:, index].astype(str).apply(lambda x : x.split(' '))
    left_arr = []
    right_arr = []
    
    #display(column)
    #iterate through column vals
    for val in column:
        if front:
            left_arr.append(val[0])
            right_arr.append(' '.join(val[1:]))
        #if not front
        else:
            left_arr.append(' '.join(val[:-1]))
            right_arr.append(val[-1])
    #make new cols and drop old
    ans_df = df.drop(df.columns[index], axis=1)
    #add right column first
    ans_df.insert(index, new_names[1], right_arr)
    ans_df.insert(index, new_names[0], left_arr)
    return ans_df


"""
Cleans the cause column and consolidates labels

params:

val: the string value for the cause

Example:
Consider the following strings:
"Playing with fire", "p/w/f", "play/w/fire"
These all point to the same cause, however we want to create a standarddized cause
to perform analysis

We then group all of these unique values and change them to a standard string:
"Playing with fire"

returns:
new standardized cause
"""
def cause_cleaner(val):
    string = str(val).lower().strip(' ').strip('.')
    #nan
    nan_str = ['na', 'nan', 'n/a', '']
    #under investigation
    ui_str = ['under invest', 'ui']
    #undetermined
    undet_str = ['undetermined', 'undet']
    #equipment
    eqt_str = ['equipment', 'equip', 'equip. use', 'welding', 'weadeater']
    #miscellaneous
    misc_str = ['misc', 'miscellaneous', 'other', 'human/other']
    #electrical
    elec_str = ['elec. power', 'elec. pow', 'powerlines', 'powerline', 'power equip', 'electrical power']
    #shooter
    shoot_str = ['shooter', 'shooting', 'munition', 'ammunition', 'shooters']
    #playing with fire
    pwf_str = ['play/w/fire', 'p/w/f', 'playing with fire']
    #transportation
    transport_str = ['railroad', 'vehicle']
    #control burned
    control_str = ['escape control', 'escape', 'control burn', 'prescribed']
    #debris
    debris_str = ['debris', 'debris burning']
    #campfire
    camp_str = ['campfire', 'rekindle']
    if string in nan_str:
        return np.nan
    elif string in ui_str:
        return 'Under Investigation'
    elif string in undet_str:
        return 'Undetermined'
    elif string in eqt_str:
        return 'Equipment'
    elif string in misc_str:
        return 'Miscellaneous'
    elif string in elec_str:
        return 'Electrical'
    elif string in shoot_str:
        return 'Guns'
    elif string in pwf_str:
        return 'Playing With Fire'
    elif string in transport_str:
        return 'Transportation'
    elif string in control_str:
        return 'Controlled Burn'
    elif string in debris_str:
        return 'Debris Burning'
    elif string in camp_str:
        return 'Camping'
    else:
        return string.title()

"""
Extracts the value from a certain cell and moves it left or right, deleting the value that was in the original cell

params:
df: dataframe to edit
r: row containing value using iloc
c: column containing value using iloc
item: the item to transpose
d: direction to move data "left" "right"

returns:
dataframe object with changes applied
"""
def transposer(df, r, c, item, d):
    frame = df.copy()
    frame.iloc[r,c] = frame.iloc[r,c].replace(item, '')
    if frame.iloc[r,c] == df.iloc[r,c]:
        frame.iloc[r,c] = df.iloc[r,c]
        return frame
    else:
        if 'left' in d:
            frame.iloc[r,c-1] = item
        elif 'right' in d:
            frame.iloc[r,c+1] = item
        else:
            raise Exception('please use left or right')
        return frame

############################################################################
#                           APPLY HELPER FUNCTIONS                         #
############################################################################

'''
APPLY HELPER FUNCTION (used in Series.apply() only)

Takes in an array with year and DOY (Day Of Year) and converts it to a datetime object.
This function is needed due to odd dating format in SQL table
'''
def datetime_convert_sql(year_day_array):
    year = int(year_day_array[0])
    #issue with nan values not converting to int values
    try:
        day = int(year_day_array[1].split('.')[0])
        ans = datetime.datetime(year, 1, 1) + datetime.timedelta(days=day-1)
    except ValueError:
        ans = np.nan
    return ans
'''
APPLY HELPER FUNCTION (used in Series.apply() only)

Takes in a string in the format MM/DD/YY and converts it to a datetime object
This function is needed due to YY format and NaN error handling
'''
def datetime_convert_pdf(date):
    date_arr = date.split('/')
    if len(date_arr) == 3:
        try:
            return datetime.datetime(int('20'+date_arr[2].split('.')[0]), int(date_arr[0]), int(date_arr[1]))
        except ValueError:
            return np.nan
    else:
        return np.nan
'''
APPLY HELPER FUNCTION (used in Series.apply() only)

Takes in a string value and converts to integer value
This function is needed for NaN value handleing to join PDF table and SQL table
'''
def int_convert(val):
    try:
        ans = int(val.replace(",", ""))
    except:
        ans = 0
    return ans

'''
APPLY HELPER FUNCTION (used in Series.apply() only)

Takes in a string and list of counties and checks if string c contains any of the counties in c_lst
'''
def county_cleaner(c, c_lst):
    #remove any dashes
    string = c.split("-")[0]
    #remove any slashes
    string = c.split("/")[0]
    #standardize
    string = string.title().strip(" ")
    #use n_grams to match to potential couny
    string_split = string.split(" ")
    for i in np.arange(len(string_split)+1):
        n_gram = " ".join(string_split[0:i]) + " County"
        if n_gram in c_lst:
            return n_gram
    return np.nan

'''
APPLY HELPER FUNCTION (used in Series.apply() only)

Takes in a string veg and cleans out invalid strings
'''
def veg_cleaner(veg):
    string = str(veg).lower().replace(' ', '').replace(',', '').replace('.', '')
    #check null
    veg_str = ['na', 'n/a', 'nan', '']
    if string in veg_str:
        return np.nan
    try:
        int(veg)
        return np.nan
    except ValueError:
        if len(veg) > 4:
            return np.nan
        
    return string.upper()

'''
APPLY HELPER FUNCTION (used in Series.apply() only)

Categorizes the burned total into 3 categories
'''
def burn_cat(val):
    if val < 100:
        return 0
    elif val < 5000:
        return 1
    else:
        return 2

'''
APPLY HELPER FUNCTION (used in Series.apply() only)

Removes numbers from county column 
'''
def resolve_county(s):
    #check if string is a number
    try:
        s_int = int(s)
        c_nm = number_county_dict[s_int]
        return c_nm
    #not a number
    except ValueError:
        c_nm = s.split('/')[0]
        c_nm = c_nm.title() + ' County'
        return c_nm

'''
APPLY HELPER FUNCTION (used in Series.apply() only)

Resolves different null value types and changes values from string to int
'''
def fatality_cleaner(val):
    string = str(val).split('.')[0]
    null_str = ['nan', 'na', 'Unnamed: 14','Unnamed: 15', 'n/a']
    if string in null_str:
        return np.nan
    return int(string)
```

```python
from scipy.stats import linregress
# Goal: FIND THE MOST PROBABLE COUNTIES THAT IS SUSCEPTIBLE TO FIRE |
# clean_df = clean_df.drop(494)
unique_counties = clean_df["COUNTY/UNIT"].dropna().unique()
years = clean_df["START_DATE"].apply(lambda x: x.year).unique()
# Removed the abnormal year data
years = np.delete(years, 10)
years = np.delete(years, 7)
res_tuples = []

for county in unique_counties:
    # Get the count of fires per year
    county_df = clean_df[clean_df["COUNTY/UNIT"] == county]
    county_df["YEAR"] = county_df["START_DATE"].apply(lambda x: x.year)
    years_total = county_df.groupby('YEAR').mean().BURNED_TOTAL
    for year in years:
        try:
            years_total[year]
        except KeyError:
            years_total[year] = 0
    
    res = linregress(years_total.index, years_total.values)
    slope = res[0]
    intercept = res[1]
    r_squared = res[2]
    p_val = res[3]
    res_tuples.append((county, slope, intercept, r_squared, p_val))

res_tuples.sort(key=lambda tup: tup[1])
# Extract the top-5
best_tuples = res_tuples[-5:]
for tup in best_tuples:
    slope = tup[1]
    intercept = tup[2]
    y_pts = []
    for year in years:
        y_pts.append( slope * year + intercept)  
    county_fireplot = sns.lineplot(years, y_pts, label=tup[0])
    
county_fireplot.set(xlabel="Year", ylabel="Predicted Average Burned Total Fire", title="Top 5 Slopes of County & Burned_Total")

display("The 5 Counties with the largest slopes:")
for tuple in best_tuples:
    res_str = "County Name: " + str(tuple[0]) + ", Slope: " + str(tuple[1]) 
    display(res_str)
```

```python
fatality_df = clean_df[['ORIGIN_DPT', 'BURNED_TOTAL', 'VEG_TYPE', 'CAUSE', 'FATALITY_FIRE', 'FATALITY_CIVIL']]
fatality_df['FATAL'] = fatality_df[['FATALITY_FIRE', 'FATALITY_CIVIL']].apply(lambda x: 1 if x.any() else 0, axis=1)

#fixing massive class imbalance using bootstrap
bootstrap_dpt = np.random.choice(fatality_df.ORIGIN_DPT.values, 900, replace=True)
bootstrap_burned_total = np.random.choice(fatality_df.ORIGIN_DPT.values, 900, replace=True)
bootstrap_veg_type = np.random.choice(fatality_df.ORIGIN_DPT.values, 900, replace=True)
bootstrap_cause = np.random.choice(fatality_df.ORIGIN_DPT.values, 900, replace=True)
bootstrap_fatal = [1] * 900

bootstrap_df = pd.DataFrame({'ORIGIN_DPT' : bootstrap_dpt,
                             'BURNED_TOTAL' : bootstrap_burned_total,
                             'VEG_TYPE' : bootstrap_veg_type,
                             'CAUSE' : bootstrap_cause,
                             'FATALITY_FIRE' : [0] * 900,
                             'FATALITY_CIVIL' : [0] * 900,
                             'FATAL' : bootstrap_fatal})

balance_df = pd.concat([fatality_df, bootstrap_df])

X = balance_df[['ORIGIN_DPT', 'BURNED_TOTAL', 'VEG_TYPE', 'CAUSE']]
y = balance_df['FATAL']
```

### Data Pipelines and Predictions

```python
#train_test_split
X_train, X_test, y_train, y_yest = train_test_split(X, y)
#deal with class imbalance

###
bayes = GaussianNB()
knn = KNeighborsClassifier()
tree = DecisionTreeClassifier()
forest = RandomForestClassifier() 
#fit
#column_transformer into pipeline
#col_transformer
categorical_cols = ["ORIGIN_DPT", "VEG_TYPE", "CAUSE"]
col_trans = ColumnTransformer([("hottie", sklearn.preprocessing.OneHotEncoder(handle_unknown= "ignore", sparse=False), categorical_cols)])

#data pipelines
#naive bayes
pipe1 = sklearn.pipeline.Pipeline([('preproc', col_trans), ('classifier', bayes)]).fit(X_train.fillna("None"), y_train)
guess_bayes = pipe1.predict(X_test.fillna("None"))
#knn
pipe2 = sklearn.pipeline.Pipeline([('preproc', col_trans), ('classifier', knn)]).fit(X_train.fillna("None"), y_train)
guess_knn = pipe2.predict(X_test.fillna("None"))
#descision tree
pipe3 = sklearn.pipeline.Pipeline([('preproc', col_trans), ('classifier', tree)]).fit(X_train.fillna("None"), y_train)
guess_tree = pipe3.predict(X_test.fillna("None"))
#random forest
pipe4 = sklearn.pipeline.Pipeline([('preproc', col_trans), ('classifier', forest)]).fit(X_train.fillna("None"), y_train)
guess_forest = pipe4.predict(X_test.fillna("None"))
```

### Model Scoring

Due to the relative rarity of our event happening, we want to place more emphasis on the *recall* score of our function. Nevertheless we don't want to ignore precision either so we will use the `F2` score to grade our models. Since `sklearn` doesn't have an `F2` score, we will make our own using the *confusion matrix* which contains **True Positives (TP)**, **False Positives (FP)**, **True Negatives (TN)**, and **False Negatives (FN)**. 

The formula for `F2` score is as follows:

<img src="http://bit.ly/317jgjQ" align="center" border="0" alt="F_{2} = \frac{5}{4} *  \frac{precision * recall}{precision + recall}   " width="348" height="71" />

```python
def f2_score(cm):
    tn = cm[0][0]
    fn = cm[1][0]
    tp = cm[1][1]
    fp = cm[0][1]
    precision = tp / (tp + fp)
    recall = tp / (tp + fn)
    f1 = precision * recall / (precision + recall)
    return (5 / 4) * f1
```

```python
bayes_score = f2_score(sklearn.metrics.confusion_matrix(guess_bayes, y_yest))
knn_score = f2_score(sklearn.metrics.confusion_matrix(guess_knn, y_yest))
tree_score = f2_score(sklearn.metrics.confusion_matrix(guess_tree, y_yest))
forest_score = f2_score(sklearn.metrics.confusion_matrix(guess_forest, y_yest))

display("F2 results found: ")
display("Bayes score: " + str(bayes_score) + ", KNN score: " + str(knn_score) + "Tree score: " + str(tree_score) + ", Forest score: " + str(forest_score))
```
