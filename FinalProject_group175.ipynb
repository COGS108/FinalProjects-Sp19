## Research Question

## Background and Prior Work

Obesity is defined as weight that is higher than what is considered a healthy weight for a given height. BMI or "Body Mass Index" is used to screen for obesity. Obesity is characterized by having a high BMI (above 30.0). Obesity is a complex issue that can be influenced by several factors ranging from genetics to behaviour in regards to diet and physical activity ("Defining...CDC"). At the root however, obesity is the result of exessive weight gain beyond a certain threshold for a given height (as per the defintion). Weight gain is the result of consumng more calories than the body needs or uses ("Maintaining... Guide").

The medical care costs of obesity in the United States was approximately $147 billion in 2008 ("Defining...CDC"). Approximately 1 in 3 adults and 1 in 6 children are obese (CDC). This is clearly a large problem that afflicts a large number of people in the United States. While it is known that obesity, at the end of the day, is a balance sheet with regards to calories, it is still an open question as to what factors cause this imbalance.

It is speculated by some that the imbalance is a result of a scarcity of affordable healthy foods. This makes poor people more susceptible to being obese (Ellen). This claim is highly suspect when considering that based on DeSilver, income and obesity aren’t proportional; in some cases obesity rates increase with income. In conjunction with a CDC report that states that only twenty three percent of Americans get enough exercise (Ducharme), it appears that a lot of Americans may be obese because they just don’t get out enough.

## Hypothesis

Obesity in America is a result of laziness and not income inequality. ### Data Cleaning

#### Overview

There are two types of data that are utilized throughout the project: state data and county data. The following guidelines were used to clean the data:
- All datasets must be stored in *pandas* dataframe objects. This requirement standardizes the container that is used and thus simplifies the use of functions from the *pandas* library.
- Dataframes must not be too wide (have too many columns so that they have to be scrolled through) and must be readable when printed. This means there must not be any incorrect/inapropriate entries and the column names must be human-readable (i.e. not obtuse abbreviations). This requirement ensures that dataframes and tables are easily readable and not confusing to the reader.
- All dataframes containing state data must have a column called **State** that contains a unique entry in that column that belongs to a set of standardized state names. This requirement simplifies merging several dataframes containing state data together and removes the need to remove incorrect entries from them afterward.
- All dataframes containing state data must contain exactly 50 rows. This requirement enforces correctness of the state data.
- All dataframes containg county data must contain a **State** and **County** columns that have unique (**State**, **County**) for every row. This requirement simplifies merging several dataframes containing county data together and removes the need to remove incorrect entries from them afterward.
- All dataframes containing county data must contain exactly 3141 rows. This requirement enforces correctness of the county data.
- Every cloumn of a dataframe should be cast to the appropriate data type (i.e columns containing strings should be cast to *str* whereas columns containg percentages or ratios should be cast to *float*). This requirement simplifies data processing and visualization.
- Columns in a dataframe should either contain names of states/counties or ratios/percentages. This requirement prevents comparisons between absolute population numbers which are bad comparisons because they are weighted towards states or counties with larger populations.
- Tests must be run on the datasets after cleaning to ensure they are correct. This ensures the accuracy of the data.

The general principle is to ensure that all columns are of the right data type and that the state and county name pairs remain consistent so that merging the dataframes do not produces any unintended consequences (like extra rows). The cleaning for most of the datasets involved reorganizing the data and sanitizing the entries (removing leading and trailign whitespace, fixing character encoding issues, etc). Some rows (especially with the dataframes containing county data) had to be manually edited for correctness since they could not be corrected algorithmically. The specifics are detailed as comments in the code.

## Ethics & Privacy
Most of the data is sourced from several federal agencies like the Center for Disease Control (CDC) and the U.S. Census Bureau with all personal identifiers scrubbed from the data. It is safe to assume that this data is safe to use and is void of bias. While the data sets can be publicly accessed, there should be no privacy concerns since all the data points are aggregates and as a result no individual or personally identifying information can be gleamed from the data. No copyrighted material was used in this project.

At least two of of the datasets used were based on surveys. These datasets include the *National Obesity by State* dataset and the *Physical Activity State Indicator Report*. The problem with surveys is that the data could potentially be skewed in either way. For example: maybe the individual does not want to admit that they are obese since it is embarrassing to admit or maybe they are very self concious and think they are obese even if they are not. As a result, it is difficult to whether the data is an overestimate or an underestimate. It is possible that with sufficiently large datasets, the participants of the survey that overestimate or underestimate their responses cancel each other out but that is a huge assumption. 

There is one major issue with the cleaning of the data that should be mentioned even though it's impact is negligible. There are 3,141 counties in the U.S excluding the District of Columbia. However, After cleaning out *Food Environment Atlas* datasets, there were 3,142 entries. This was problematic since this was not supposed to be the case. There was one unaccounted county in the dataset. The county shapefile on the other hand had 3,141 entries after being cleaned. Based on the intersection of these two datasets, it was revealed that Bedford City County was the 3,142nd entry. This was summarily dropped from the *Food Environment Atlas* datasets however it is difficult to say whether that was they right county to drop since what if the county shapefile was incorrect? In summary, the may potentially be one county in the *Food Environment Atlas* datasets that may not exist in reality.

The results from the analysis are provided to be used for educational and research purposes. They can be used by local governments to improve general welfare of citizens by providing physical activity programs or easily accessible sources of healthy and affordable foods whereever needed. However, there is the possibility that the analysis could somehow be utilized by corporations looking to exploit locations stricken by poverty. Forms of this malicious intent include building new fast food restaurants in poverished parts of cities. While not much can be done to prevent this type of behavior, the purpose of this analysis is not meant to harm or bias any minority or other groups.

In the case that the analysis proves to be prominent in the harm of certain groups, the project can be set to private on Github to prevent distribution and viewing.
