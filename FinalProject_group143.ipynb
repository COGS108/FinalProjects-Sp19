## Data Cleaning For Population 2000 Sheet

#### This data was provided by the Census Bureau. This data had a little more than we needed, so more cleaning was done. We transposed the data, as the Bureau organized it differently. We then deleted some unwanted rows. We also noticed that values were given in percentages and whole numbers, luckily they were separated in an 'every other' fashion, so we deleted every other row to save the whole number values. We then deleted any values below 10,000 as we felt that was not going to play a large role in the majority. Lastly, we made some cosmetic changes with some dropped rows and renaming for better readability. 

```python
race_2000df = pd.read_excel('populationtotals2000.xlsm')
race_2000df = race_2000df.T
race_2000df.columns = ['Race', 'Percent/Total Number']
race_2000df = race_2000df.iloc[3:] # deleted first 3 rows
race_2000df = race_2000df.iloc[::2] # deleted all percentage values
race_2000df = race_2000df.iloc[2:] # delete total population row
race_2000df['Percent/Total Number'] = race_2000df['Percent/Total Number'].astype(int)


race_2000df2 = race_2000df[race_2000df['Percent/Total Number'] > 10000] 
race_2000df2 = race_2000df2.drop(['HC01_VC07','HC01_VC12','HC01_VC13','HC01_VC14','HC01_VC15','HC01_VC16','HC01_VC18','HC01_VC17','HC01_VC35','HC01_VC38','HC01_VC39','HC01_VC41','HC01_VC42','HC01_VC43','HC01_VC44','HC01_VC45','HC01_VC46','HC01_VC47' ,'HC01_VC33','HC01_VC36'])
race_2000df3 = race_2000df2.replace(['Number; RACE - Total population - One race - White','Number; RACE - Total population - One race - Black or African American','Number; RACE - Total population - One race - American Indian and Alaska Native','Number; RACE - Total population - One race - Asian','Number; HISPANIC OR LATINO - Total population','Number; RACE - Total population - One race - Native Hawaiian and Other Pacific Islander','Number; HISPANIC OR LATINO - Total population - Hispanic or Latino (of any race)'],['White','African American/Black','American Indian','Asian', 'Hispanis/Latino','Native Hawaiian','Hispanic/Latino']) 
race_2000df3.set_index('Race', inplace=True)

race_2000df3.style
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
from sklearn.decomposition import PCA
import pandas as pd
from sklearn.metrics import explained_variance_score
from sklearn.preprocessing import StandardScaler
from sklearn import linear_model
from scipy import stats
from functools import reduce
import seaborn as sn
from sklearn.metrics import confusion_matrix

df_Age = pd.read_excel('Age.xlsx', index_col = 0)
df_Job = pd.read_excel('Job.xlsx', index_col = 0)
df_Income = pd.read_excel('Income.xlsx', index_col = 0)
df_Population = pd.read_excel('Population.xlsx', index_col = 0)
df_Price = pd.read_excel('Price.xls', index_col = 0)

df_Age.index = df_Age.index.str.lower()
df_Job.index = df_Job.index.str.lower()
df_Income.index = df_Income.index.str.lower()
df_Population.index = df_Population.index.str.lower()
df_Price.index = df_Price.index.str.lower()

cities = ['carlsbad', 'chula vista', 'coronado', 'del mar', 'el cajon', 'encinitas', 'escondido', 'imperial beach',
          'la mesa', 'lemon grove', 'national city', 'oceanside', 'poway', 'san dieog', 'san marcos', 'santee', 'solana beach', 'vista']

df_Age = df_Age.loc[cities]
df_Job = df_Job.loc[cities]
df_Income = df_Income.loc[cities]
df_Population = df_Population.loc[cities]
df_Price = df_Price.loc[cities]

dfs = [df_Age, df_Job, df_Income, df_Population]
df_X =  reduce(lambda left,right: pd.merge(left, right, left_index = True, right_index = True), dfs)

df_X = df_X.apply(lambda x: x.fillna(x.mean()),axis = 0) # Filling NaN values with the mean
df_Price = df_Price.apply(lambda x: x.fillna(x.mean()),axis = 0)

ss = StandardScaler()
df_X= ss.fit_transform(df_X)
df_X = pd.DataFrame(df_X)

number_train = int (0.7 * len(df_X))

train_x, train_y = df_X.iloc[:number_train], df_Price.iloc[:number_train]
test_x, test_y = df_X.iloc[number_train:], df_Price.iloc[number_train:]


regr = linear_model.Lasso(alpha = 9, normalize= False)
regr.fit(train_x,train_y)

train_perf = r2_score(train_y, regr.predict(train_x))
test_perf = r2_score(test_y, regr.predict(test_x))

train_MSE = mean_squared_error(train_y, regr.predict(train_x))
test_MSE = mean_squared_error(test_y, regr.predict(test_x))

print("MSE of Train data is %.4f" %(train_MSE))
print("MSE of test data is %.4f" %(test_MSE))
print("R2 of Train data is %.4f" %(train_perf))
print("R2 of Test data is %.4f" %(test_perf))

regr.coef_
```

#### We performed a Lasso Regression using our data separated by cities. We sorted the data by the list called 'cities' first, then we merged them into one DataFrame. And since there were some cities with NaN values in some columns, we filled them up with there mean value. Later we scaled the data using StandardScaler. Finally, we tried to find the optinal alpha value for Lasso Regression and it turned out to be alpha = 9. # Ethics and Privacy

#### The main use of our data is to understand how people look for ideal housing around the San Diego County. Although not perfect, we strive to create a projection that can show where a certain demographic or population may live in. To do this, we take in consideration of race, age, and income. The target group we aim for are usually students or new families trying to find residency around San Diego. A side effect of this categorization of demographics can lead to stigmatization. Stigmatization is the act of categorizing and giving a title to a person or group that leads to stereotyping or other connotations. Some people may not want to be categorized by race or income or even be labeled so this can be a potential ethical issue. One way to reduce this effect is to make sure we clarify that not everyone living in the area is the specific income, race, or age. The projection is based off of the average or the majority of the people living there.
#### Another ethical issue that can be brought up is that some people are protective of how much they are paying for their living space. Many people consider their living area or house to be a private area or space, and would not like other people to see their information. Furthermore, observations toward these details might not be warranted. This again can be solved with the emphasis on the average of the income. The projection does not show the exact rent/payment for each household so there is privacy for the individual households.
