```python
import pandas as pd
import numpy as np
```

```python
#Main Dataframes
df_orders = pd.read_csv('../instacart-market-basket-analysis/orders.csv')
df_prod_p = pd.read_csv('../instacart-market-basket-analysis/order_products__prior.csv')
```

```python
def group_orders_by_uid(uid):
    associated_oid = list(df_orders.loc[(df_orders['user_id'] == uid) & (df_orders['eval_set'] == 'prior')]['order_id'])
    master = pd.DataFrame()
    for oid in associated_oid:
        master = master.append(df_prod_p.loc[df_prod_p['order_id'] == oid])
    return master

def user_data():
    user_count = dict()
    for i, row in df_orders.iterrows():
        uid = df_orders.loc[i, 'user_id']
        if user_count.get(uid) == None:
            user_count[uid] = 1
        else:
            user_count[uid] = (user_count[uid] + 1)
    return user_count
```

```python
user_data_dict = user_data()
```

```python
pd.DataFrame(user_data_dict,columns=["user_id","number of orders placed"])
```
